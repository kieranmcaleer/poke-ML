```python
import pandas as pd 
```

```python
poke_csv="Resources/clean_data.csv"
poke_df= pd.read_csv(poke_csv)
poke_df
```

```python
short_df=poke_df.drop(["Total", "HP", "Attack", "Defense", "SpAtk", "SpDef", "Speed", "Legendary"], axis=1)
short_df
```

```python
#Clean the dataframe to include only relevant columns and create a recode column for each type 
short_df["Type1_Recode"] = ""
short_df["Type2_Recode"] = ""
short_df
```

```python
#Gather list of unique pokemon types
type1_list = short_df["Type1"].unique()
type1_list
```

```python
#Confirm number of unique types
len(type1_list)
```

```python
#iterate through DF to recode the listed type1 for each pokemon 
for row,col in short_df.iterrows(): 
    if short_df["Type1"][row] == "Grass": 
        short_df["Type1_Recode"][row] = 1
    elif short_df["Type1"][row] == "Fire": 
        short_df["Type1_Recode"][row] = 2
    elif short_df["Type1"][row] == "Water": 
        short_df["Type1_Recode"][row] = 3
    elif short_df["Type1"][row] == "Bug": 
        short_df["Type1_Recode"][row] = 4
    elif short_df["Type1"][row] == "Normal": 
        short_df["Type1_Recode"][row] = 5
    elif short_df["Type1"][row] == "Poison": 
        short_df["Type1_Recode"][row] = 6
    elif short_df["Type1"][row] == "Electric": 
        short_df["Type1_Recode"][row] = 7
    elif short_df["Type1"][row] == "Ground": 
        short_df["Type1_Recode"][row] = 8
    elif short_df["Type1"][row] == "Fairy": 
        short_df["Type1_Recode"][row] = 9
    elif short_df["Type1"][row] == "Fighting": 
        short_df["Type1_Recode"][row] = 10
    elif short_df["Type1"][row] == "Psychic": 
        short_df["Type1_Recode"][row] = 11
    elif short_df["Type1"][row] == "Rock": 
        short_df["Type1_Recode"][row] = 12
    elif short_df["Type1"][row] == "Ghost": 
        short_df["Type1_Recode"][row] = 13
    elif short_df["Type1"][row] == "Ice": 
        short_df["Type1_Recode"][row] = 14
    elif short_df["Type1"][row] == "Dragon": 
        short_df["Type1_Recode"][row] = 15
    elif short_df["Type1"][row] == "Dark": 
        short_df["Type1_Recode"][row] = 16
    elif short_df["Type1"][row] == "Steel": 
        short_df["Type1_Recode"][row] = 17
    elif short_df["Type1"][row] == "Flying": 
        short_df["Type1_Recode"][row] = 18
        
short_df
```
